```
import plotly as py
import plotly.graph_objs as go
import pandas as pd
```

```
tesla = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/tesla-stock-price.csv")
```

```
trace_one = go.Scatter(x=tesla.date,y=tesla['open'],name="Tesla Open",line=dict(color='#17BECF'),opacity=0.8)
trace_two = go.Scatter(x=tesla.date,y=tesla['close'],name="Tesla Close",line=dict(color='#7F7F7F'),opacity=0.8)
```

```
data = [trace_one,trace_two]
layout = dict(title = 'Tesla Stock Price - Open vs Close')
#fig = dict(data = data, layout= layout)
#iplot(fig, filename = 'Tesla Stock Comaparison')
fig = go.Figure(data=data,layout=layout)
fig.show()
```
